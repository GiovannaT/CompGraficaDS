```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
file = 'sismigraPF.xlsx'
dataframe = pd.read_excel(file, engine='openpyxl').fillna(value=0)
dataframe
```

```python
dataframeParaiba = dataframe[dataframe.UF.eq('PB')]
dataframeParaiba
```

```python
dataframeParaiba.rename(columns = {'QTD,': 'QTD'}, inplace = True)
dataframeParaiba
```

```python
dataframeParaibaTotal = dataframeParaiba.QTD.sum()
dataframeParaibaTotal
```

```python
dataframeParaibaResid = dataframeParaiba.sort_values(['QTD,'],ascending=False)
dataframeParaibaResid
```

```python
dataframePaisesClassification = dataframeParaiba['CLASSIFICACAO'].value_counts()
dataframePaisesClassification
```

```python
classification = {'type': ['Residente', 'Temporário', 'Provisório'], 'quantity': [101, 45, 9]}
df = pd.DataFrame(data=classification)
df
```

```python
labels = df['type']
valores = df['quantity']
plt.figure(figsize=(7,7))
plt.pie(x=valores, labels=labels, autopct='%1.1f%%')
plt.show()
```
